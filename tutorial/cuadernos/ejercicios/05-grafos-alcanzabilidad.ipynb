# Ejercicios: grafos y alcanzabilidad

Este cuaderno practica representación de grafos y búsqueda de caminos.

La pregunta principal será: ¿existe un camino desde un vértice inicial hasta un vértice objetivo?

```python
from collections import deque


grafo = {
    "A": ["B", "C"],
    "B": ["D"],
    "C": ["E"],
    "D": ["F"],
    "E": ["D", "G"],
    "F": [],
    "G": ["F"],
}


def alcanzable_bfs(grafo, inicio, objetivo):
    visitados = set()
    cola = deque([inicio])
    while cola:
        actual = cola.popleft()
        if actual == objetivo:
            return True
        if actual in visitados:
            continue
        visitados.add(actual)
        cola.extend(grafo.get(actual, []))
    return False


def alcanzable_dfs(grafo, inicio, objetivo):
    visitados = set()
    pila = [inicio]
    while pila:
        actual = pila.pop()
        if actual == objetivo:
            return True
        if actual in visitados:
            continue
        visitados.add(actual)
        pila.extend(grafo.get(actual, []))
    return False
```

## Ejercicio 1: consultas de alcanzabilidad

Prueba varios pares de vértices.

```python
consultas = [("A", "F"), ("C", "B"), ("E", "F"), ("B", "G")]

for inicio, objetivo in consultas:
    print(f"{inicio} -> {objetivo}: BFS={alcanzable_bfs(grafo, inicio, objetivo)} DFS={alcanzable_dfs(grafo, inicio, objetivo)}")
```

## Ejercicio 2: reconstruir un camino

Ahora no solo queremos saber si existe camino, sino recuperar uno.

```python
def camino_bfs(grafo, inicio, objetivo):
    cola = deque([(inicio, [inicio])])
    visitados = set()
    while cola:
        actual, camino = cola.popleft()
        if actual == objetivo:
            return camino
        if actual in visitados:
            continue
        visitados.add(actual)
        for vecino in grafo.get(actual, []):
            cola.append((vecino, camino + [vecino]))
    return None


print(camino_bfs(grafo, "A", "F"))
print(camino_bfs(grafo, "C", "B"))
```

## Ejercicio 3: modificar el grafo

Añade una arista de `C` a `B` y observa cómo cambia la consulta `C -> B`.

```python
grafo_modificado = {vertice: vecinos[:] for vertice, vecinos in grafo.items()}
grafo_modificado["C"].append("B")

print("antes:", alcanzable_bfs(grafo, "C", "B"))
print("después:", alcanzable_bfs(grafo_modificado, "C", "B"))
print("camino:", camino_bfs(grafo_modificado, "C", "B"))
```
