# Ejercicios: selección de modelos con entropía cruzada

Este cuaderno compara modelos probabilísticos candidatos usando entropía cruzada empírica.

La pregunta práctica es: ¿qué modelo asigna mayor probabilidad a los datos observados?

```python
import math
from collections import Counter


def distribucion_empirica(datos):
    total = len(datos)
    return {simbolo: cuenta / total for simbolo, cuenta in Counter(datos).items()}


def entropia_cruzada_empirica(datos, modelo):
    total = 0
    for simbolo in datos:
        p = modelo.get(simbolo, 0)
        if p == 0:
            return math.inf
        total -= math.log2(p)
    return total / len(datos)


def entropia(distribucion):
    return -sum(p * math.log2(p) for p in distribucion.values() if p > 0)
```

## Datos observados

Tenemos una secuencia de símbolos y varios modelos candidatos.

```python
datos = list("AAAAABBBBCCDAAABBBCCAAAAB")
empirica = distribucion_empirica(datos)

modelos = {
    "uniforme": {"A": 0.25, "B": 0.25, "C": 0.25, "D": 0.25},
    "sesgado A": {"A": 0.5, "B": 0.25, "C": 0.15, "D": 0.10},
    "muy sesgado": {"A": 0.8, "B": 0.1, "C": 0.05, "D": 0.05},
    "sin D": {"A": 0.5, "B": 0.3, "C": 0.2, "D": 0.0},
}

print("Distribución empírica:", empirica)
print(f"H(empírica) = {entropia(empirica):.4f} bits/símbolo")

for nombre, modelo in modelos.items():
    h = entropia_cruzada_empirica(datos, modelo)
    print(f"{nombre:12s} H_emp(P,Q) = {h}")
```

## Elegir el mejor modelo

Menor entropía cruzada significa mejor ajuste a estos datos, en bits por símbolo.

```python
resultados = []
for nombre, modelo in modelos.items():
    resultados.append((entropia_cruzada_empirica(datos, modelo), nombre))

for h, nombre in sorted(resultados):
    print(f"{nombre:12s} {h}")
```
