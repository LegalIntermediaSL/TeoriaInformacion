## Preparación

Estas funciones sirven como base para todos los ejercicios.

```python
import math


def informacion_propia(p):
    if p <= 0 or p > 1:
        raise ValueError("p debe estar en (0, 1].")
    return -math.log2(p)


def entropia(distribucion):
    total = sum(distribucion.values())
    if not math.isclose(total, 1.0, abs_tol=1e-9):
        raise ValueError(f"La distribución debe sumar 1; suma {total}.")
    return sum(p * informacion_propia(p) for p in distribucion.values() if p > 0)
```

## Ejercicio 1: sorpresa

Calcula la información propia de resultados con probabilidades `1`, `1/2`, `1/4`, `1/8` y `1/16`. ¿Qué patrón observas?

```python
probabilidades = [1, 1/2, 1/4, 1/8, 1/16]

for p in probabilidades:
    print(f"p = {p:>6.4f}  I(x) = {informacion_propia(p):.2f} bits")
```

## Ejercicio 2: ordenar fuentes por entropía

Ejecuta la celda y luego añade una fuente nueva. Intenta predecir su posición antes de calcularla.

```python
fuentes = {
    "segura": {"A": 1.0},
    "moneda equilibrada": {"A": 0.5, "B": 0.5},
    "moneda sesgada": {"A": 0.9, "B": 0.1},
    "cuatro uniformes": {"A": 0.25, "B": 0.25, "C": 0.25, "D": 0.25},
    "cuatro concentrada": {"A": 0.7, "B": 0.1, "C": 0.1, "D": 0.1},
}

ordenadas = sorted(fuentes.items(), key=lambda item: entropia(item[1]))
for nombre, distribucion in ordenadas:
    print(f"{nombre:20s} H = {entropia(distribucion):.4f} bits")
```

## Ejercicio 3: entropía máxima

Para una fuente uniforme de `N` símbolos, la entropía debe ser `log2(N)`. Compruébalo para varios valores de `N`.

```python
def uniforme(n):
    return {f"s{i}": 1 / n for i in range(n)}


for n in [2, 4, 8, 16, 32]:
    h = entropia(uniforme(n))
    esperado = math.log2(n)
    print(f"N = {n:2d}  H = {h:.2f}  log2(N) = {esperado:.2f}")
```

## Preguntas de cierre

1. ¿Por qué una fuente segura tiene entropía cero?
2. ¿Por qué una fuente uniforme maximiza la entropía?
3. ¿Qué cambia si aumentamos el número de símbolos posibles?

```python
# ── Verificación automática ──────────────────────────────────────────────────
import math

def H(p): return -sum(pi * math.log2(pi) for pi in p if pi > 0)

# Entropía máxima es log2(n) para distribución uniforme
assert abs(H([0.25, 0.25, 0.25, 0.25]) - 2.0) < 1e-9, "H uniforme n=4 debe ser 2 bits"
# Entropía mínima es 0 para distribución determinista
assert H([1.0]) == 0.0, "H de certeza debe ser 0"
# Entropía de Bernoulli en p=0.5 es 1 bit
assert abs(H([0.5, 0.5]) - 1.0) < 1e-9, "H Bernoulli(0.5) debe ser 1 bit"
print("✓ Todas las verificaciones de entropía pasan")
```
