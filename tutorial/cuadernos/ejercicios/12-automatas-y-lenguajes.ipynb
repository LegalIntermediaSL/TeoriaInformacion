# Ejercicios: autómatas y lenguajes formales

Diseña autómatas finitos deterministas (DFA) y verifica su funcionamiento con casos de prueba. Cada ejercicio propone un lenguaje y pide construir el DFA que lo reconozca.

**Artículos asociados:** [Autómatas finitos y lenguajes regulares](../../03-computabilidad/05-automatas-finitos-y-lenguajes-regulares.md), [Gramáticas y la jerarquía de Chomsky](../../03-computabilidad/06-gramaticas-y-jerarquia-chomsky.md)

```python
class DFA:
    """Autómata finito determinista."""

    def __init__(self, states, alphabet, transitions, initial, accepting):
        """
        states      : conjunto de estados
        alphabet    : conjunto de símbolos
        transitions : dict {(estado, símbolo): nuevo_estado}
        initial     : estado inicial
        accepting   : conjunto de estados de aceptación
        """
        self.states = set(states)
        self.alphabet = set(alphabet)
        self.transitions = transitions
        self.initial = initial
        self.accepting = set(accepting)

    def run(self, string):
        """Devuelve True si el DFA acepta string."""
        state = self.initial
        for symbol in string:
            key = (state, symbol)
            if key not in self.transitions:
                return False  # transición no definida → rechazar
            state = self.transitions[key]
        return state in self.accepting

    def trace(self, string):
        """Muestra la secuencia de estados al procesar string."""
        state = self.initial
        print(f'  Inicio: {state}')
        for i, symbol in enumerate(string):
            key = (state, symbol)
            if key not in self.transitions:
                print(f'  No hay transición desde {state} con "{symbol}" → rechazar')
                return
            state = self.transitions[key]
            print(f'  Símbolo "{symbol}" → {state}')
        resultado = 'ACEPTA' if state in self.accepting else 'RECHAZA'
        print(f'  Estado final: {state} → {resultado}')

def verificar_dfa(dfa, casos):
    todos_ok = True
    for entrada, esperado in casos:
        resultado = dfa.run(entrada)
        ok = '✓' if resultado == esperado else '✗'
        print(f'  {ok} entrada={repr(entrada):15} esperado={esperado} obtenido={resultado}')
        if resultado != esperado:
            todos_ok = False
    print('Todos los casos correctos.' if todos_ok else 'Hay errores.')

print('DFA y utilidades cargados.')
```

## Ejercicio 3: el pumping lemma en acción

El pumping lemma afirma que si $L$ es regular con constante de bombeo $p$, cualquier cadena $w \in L$ con $|w| \geq p$ puede descomponerse como $w = xyz$ con $|xy| \leq p$, $|y| \geq 1$ y $xy^iz \in L$ para todo $i \geq 0$.

Vamos a verificar empíricamente que el DFA para "termina en `01`" satisface la propiedad de bombeo (pero con $p =$ número de estados), y a encontrar cadenas del lenguaje $\{a^n b^n\}$ que violan el pumping lemma para lenguajes regulares.

```python
def verificar_pumping(dfa, w, p):
    """Verifica la propiedad de bombeo para una cadena w con constante p."""
    if len(w) < p:
        print(f'La cadena {repr(w)} es más corta que p={p}, no aplica.')
        return
    print(f'Cadena: {repr(w)} (longitud {len(w)})')
    # Probar todas las descomposiciones xyz con |xy| <= p, |y| >= 1
    bombeado_ok = False
    for j in range(1, p+1):       # |xy| = j, entonces x = w[:i], y = w[i:j], z = w[j:]
        for i in range(0, j):     # |x| = i, |y| = j-i >= 1
            x = w[:i]
            y = w[i:j]
            z = w[j:]
            assert len(y) >= 1
            # Comprobar que xy^i z está en el lenguaje para i=0,1,2,3
            resultados = [dfa.run(x + y*k + z) for k in range(4)]
            if all(resultados):
                bombeado_ok = True
                break
        if bombeado_ok:
            break
    if bombeado_ok:
        print(f'  Existe una descomposición válida: x={repr(x)}, y={repr(y)}, z={repr(z)}')
    else:
        print(f'  No existe descomposición válida → el lenguaje NO es regular (si w ∈ L).')

print('Verificando pumping para DFA "termina en 01" (p=3):')
for w in ['01', '001', '10101', '00001']:
    verificar_pumping(dfa_ends01, w, p=3)
    print()

print('Intentando bombear "aabb" como si perteneciera a un lenguaje regular:')
# Construimos un DFA que acepta a^n b^n solo hasta n=3 (finito = regular)
# Esto ilustra la diferencia entre el lenguaje finito y el infinito
print('(Si {a^n b^n} fuera regular, deberíamos poder bombear "a^p b^p" para p = nº estados)')
```

## Reflexión final

- Un DFA necesita tantos estados como "cosas distintas que hay que recordar" del prefijo leído.
- Para "termina en `01`": 3 estados, uno por cada sufijo relevante (vacío/otro, `0`, `01`).
- Para "número de `a` mod 3": 3 estados, uno por cada residuo.
- El lenguaje $\{a^n b^n\}$ requeriría recordar el valor exacto de $n$, lo que necesita infinitos estados → no es regular.
