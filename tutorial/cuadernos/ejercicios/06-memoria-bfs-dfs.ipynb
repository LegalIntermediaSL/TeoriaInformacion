# Ejercicios: memoria en BFS y DFS

Este cuaderno conecta búsqueda en grafos con complejidad espacial.

Mediremos el tamaño máximo de la frontera que mantienen BFS y DFS al explorar árboles regulares.

```python
from collections import deque


def hijos(nodo, factor_ramificacion, profundidad_maxima):
    valor, profundidad = nodo
    if profundidad >= profundidad_maxima:
        return []
    return [(valor * factor_ramificacion + i, profundidad + 1)
            for i in range(factor_ramificacion)]


def medir_bfs(factor_ramificacion, profundidad_maxima):
    cola = deque([(1, 0)])
    max_frontera = len(cola)
    visitados = 0
    while cola:
        nodo = cola.popleft()
        visitados += 1
        cola.extend(hijos(nodo, factor_ramificacion, profundidad_maxima))
        max_frontera = max(max_frontera, len(cola))
    return visitados, max_frontera


def medir_dfs(factor_ramificacion, profundidad_maxima):
    pila = [(1, 0)]
    max_frontera = len(pila)
    visitados = 0
    while pila:
        nodo = pila.pop()
        visitados += 1
        pila.extend(hijos(nodo, factor_ramificacion, profundidad_maxima))
        max_frontera = max(max_frontera, len(pila))
    return visitados, max_frontera
```

## Comparación por profundidad

BFS tiende a guardar muchos nodos del mismo nivel. DFS mantiene una pila más relacionada con la profundidad.

```python
factor = 3

print(f"Factor de ramificación: {factor}")
print(f"{'prof':>4} {'visitados':>10} {'max BFS':>10} {'max DFS':>10}")
for profundidad in range(1, 8):
    visitados_bfs, max_bfs = medir_bfs(factor, profundidad)
    visitados_dfs, max_dfs = medir_dfs(factor, profundidad)
    assert visitados_bfs == visitados_dfs
    print(f"{profundidad:4d} {visitados_bfs:10d} {max_bfs:10d} {max_dfs:10d}")
```

## Cambiar el factor de ramificación

A mayor número de hijos por nodo, la frontera de BFS crece muy rápido.

```python
profundidad = 6

print(f"Profundidad: {profundidad}")
print(f"{'factor':>6} {'visitados':>10} {'max BFS':>10} {'max DFS':>10}")
for factor in [2, 3, 4, 5]:
    visitados_bfs, max_bfs = medir_bfs(factor, profundidad)
    visitados_dfs, max_dfs = medir_dfs(factor, profundidad)
    print(f"{factor:6d} {visitados_bfs:10d} {max_bfs:10d} {max_dfs:10d}")
```
