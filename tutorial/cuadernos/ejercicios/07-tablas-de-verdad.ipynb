# Ejercicios: tablas de verdad

Este cuaderno genera tablas de verdad para fórmulas booleanas pequeñas.

Sirve como puente entre lógica proposicional, SAT y circuitos booleanos.

```python
from itertools import product


def asignaciones(variables):
    for valores in product([False, True], repeat=len(variables)):
        yield dict(zip(variables, valores))


def tabla(variables, formula):
    for asignacion in asignaciones(variables):
        yield asignacion, formula(asignacion)


def mostrar_tabla(variables, formula):
    print(" ".join(f"{v:>5s}" for v in variables) + " | valor")
    for asignacion, valor in tabla(variables, formula):
        fila = " ".join(f"{str(asignacion[v]):>5s}" for v in variables)
        print(f"{fila} | {valor}")
```

## Ejercicio 1: conjunción y disyunción

Compara `x and y` con `x or y`.

```python
print("x and y")
mostrar_tabla(["x", "y"], lambda a: a["x"] and a["y"])

print("\nx or y")
mostrar_tabla(["x", "y"], lambda a: a["x"] or a["y"])
```

## Ejercicio 2: De Morgan

Verifica que `not (x and y)` equivale a `(not x) or (not y)`.

```python
def izquierda(a):
    return not (a["x"] and a["y"])


def derecha(a):
    return (not a["x"]) or (not a["y"])


for asignacion in asignaciones(["x", "y"]):
    print(asignacion, izquierda(asignacion), derecha(asignacion), izquierda(asignacion) == derecha(asignacion))
```

## Ejercicio 3: una fórmula SAT pequeña

Busca asignaciones que satisfagan `(x or y) and ((not x) or z)`.

```python
def formula_sat(a):
    return (a["x"] or a["y"]) and ((not a["x"]) or a["z"])


for asignacion, valor in tabla(["x", "y", "z"], formula_sat):
    if valor:
        print("satisface:", asignacion)
```

1. Cambia la fórmula SAT y cuenta cuántas asignaciones la satisfacen.
2. Añade una cuarta variable y observa cómo se duplica el tamaño de la tabla.
3. Escribe una fórmula que sea falsa para toda asignación.

```python
# ── Verificación automática ──────────────────────────────────────────────────

def tabla_verdad(f, n_vars):
    return [f(*[(bits >> i) & 1 for i in range(n_vars)]) for bits in range(1 << n_vars)]

# AND: solo True cuando todos son 1
assert tabla_verdad(lambda a, b: a and b, 2) == [False, False, False, True]
# OR: False solo cuando todos son 0
assert tabla_verdad(lambda a, b: a or b, 2) == [False, True, True, True]
# Tautología: A ∨ ¬A
assert all(tabla_verdad(lambda a: bool(a) or not bool(a), 1)), "A v -A debe ser tautologia"
# Contradiccion: A ∧ ¬A
assert not any(tabla_verdad(lambda a: bool(a) and not bool(a), 1)), "A ^ -A debe ser contradiccion"
print("✓ Tablas de verdad correctas")
```
