# Ejercicios: códigos prefijo

Este cuaderno propone prácticas sobre códigos prefijo, longitud media y decodificación.

La idea es experimentar con códigos antes de estudiar algoritmos como Huffman.

```python
def es_prefijo(codigo):
    palabras = list(codigo.values())
    return all(i == j or not otra.startswith(palabra)
               for i, palabra in enumerate(palabras)
               for j, otra in enumerate(palabras))


def kraft(codigo):
    return sum(2 ** (-len(palabra)) for palabra in codigo.values())


def longitud_media(distribucion, codigo):
    return sum(distribucion[s] * len(codigo[s]) for s in distribucion)
```

## Ejercicio 1: comparar códigos

Ejecuta la celda y decide cuál código conviene para la fuente dada.

```python
fuente = {"A": 0.4, "B": 0.3, "C": 0.2, "D": 0.1}

codigos = {
    "fijo": {"A": "00", "B": "01", "C": "10", "D": "11"},
    "variable 1": {"A": "0", "B": "10", "C": "110", "D": "111"},
    "variable 2": {"A": "00", "B": "01", "C": "10", "D": "110"},
}

for nombre, codigo in codigos.items():
    print(nombre)
    print("  prefijo:", es_prefijo(codigo))
    print("  Kraft:", round(kraft(codigo), 4))
    if es_prefijo(codigo):
        print("  longitud media:", round(longitud_media(fuente, codigo), 4))
```

## Ejercicio 2: diseñar un código

Modifica `mi_codigo` para intentar bajar la longitud media sin romper la propiedad prefijo.

```python
mi_codigo = {"A": "0", "B": "10", "C": "110", "D": "111"}

print("prefijo:", es_prefijo(mi_codigo))
print("Kraft:", kraft(mi_codigo))
print("longitud media:", longitud_media(fuente, mi_codigo))
```

## Ejercicio 3: decodificar

Completa o modifica el mensaje codificado y comprueba que se decodifica de forma única.

```python
def decodificar(bits, codigo):
    inverso = {palabra: simbolo for simbolo, palabra in codigo.items()}
    salida = []
    buffer = ""
    for bit in bits:
        buffer += bit
        if buffer in inverso:
            salida.append(inverso[buffer])
            buffer = ""
    if buffer:
        raise ValueError("Sobra un fragmento no decodificado.")
    return salida


bits = "0101101110"
print(decodificar(bits, mi_codigo))
```

## Preguntas de cierre

1. ¿Puede un código tener suma de Kraft menor que 1 y seguir siendo útil?
2. ¿Por qué no basta con que las palabras sean distintas?
3. ¿Qué símbolo debería recibir la palabra más corta si la fuente no es uniforme?

```python
# ── Verificación automática ──────────────────────────────────────────────────

def kraft(codigos):
    return sum(2**(-len(c)) for c in codigos)

# Kraft estricto: código prefijo válido
assert kraft(['0', '10', '110', '111']) <= 1.0, "Violación de Kraft"
# Longitud media
def longitud_media(probs, codigos):
    return sum(p * len(c) for p, c in zip(probs, codigos))

p = [0.5, 0.25, 0.125, 0.125]
huffman = ['0', '10', '110', '111']
L = longitud_media(p, huffman)
import math
H = -sum(pi * math.log2(pi) for pi in p)
assert L >= H - 1e-9, "Longitud media no puede ser menor que la entropia"
assert L <= H + 1, "Longitud media Huffman no puede exceder H+1"
print(f"✓ Longitud media={L:.3f}, H={H:.3f} — código Huffman válido")
```
