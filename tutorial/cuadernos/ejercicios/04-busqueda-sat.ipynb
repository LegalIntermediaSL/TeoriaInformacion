# Ejercicios: búsqueda exhaustiva para SAT

Este cuaderno muestra cómo verificar asignaciones para fórmulas CNF y cómo una búsqueda exhaustiva crece con el número de variables.

SAT está en NP porque una asignación se verifica rápido. La dificultad aparece al buscar una asignación satisfactoria sin saber cuál probar.

```python
from itertools import product


def variables(formula):
    resultado = set()
    for clausula in formula:
        for literal in clausula:
            resultado.add(literal[1:] if literal.startswith("¬") else literal)
    return sorted(resultado)


def evaluar_literal(literal, asignacion):
    if literal.startswith("¬"):
        return not asignacion[literal[1:]]
    return asignacion[literal]


def satisface(formula, asignacion):
    return all(any(evaluar_literal(literal, asignacion) for literal in clausula)
               for clausula in formula)


def buscar_asignacion(formula):
    vs = variables(formula)
    for valores in product([False, True], repeat=len(vs)):
        asignacion = dict(zip(vs, valores))
        if satisface(formula, asignacion):
            return asignacion
    return None
```

## Ejercicio 1: verificar y buscar

La fórmula está en CNF: lista de cláusulas, cada cláusula como lista de literales.

```python
formula = [
    ["x", "y", "¬z"],
    ["¬x", "z"],
    ["¬y", "z"],
]

candidata = {"x": True, "y": True, "z": True}

print("variables:", variables(formula))
print("candidata satisface:", satisface(formula, candidata))
print("primera asignación encontrada:", buscar_asignacion(formula))
```

## Ejercicio 2: fórmula insatisfacible

Una fórmula puede exigir una variable y su negación a la vez.

```python
insatisfacible = [["x"], ["¬x"]]

print("asignación encontrada:", buscar_asignacion(insatisfacible))
```

## Ejercicio 3: crecimiento de asignaciones

Cada variable duplica el número de asignaciones posibles.

```python
for n in [5, 10, 20, 30, 40]:
    print(f"{n:2d} variables -> {2**n:>15,} asignaciones")
```
