# Ejercicios: medición de complejidad temporal

Este cuaderno mide tiempos de funciones simples. Las mediciones no sustituyen el análisis asintótico, pero ayudan a construir intuición.

Usaremos tamaños moderados para mantener el cuaderno rápido.

```python
import time


def lineal(n):
    total = 0
    for i in range(n):
        total += i
    return total


def cuadratica(n):
    total = 0
    for i in range(n):
        for j in range(n):
            total += (i + j) % 7
    return total


def medir(funcion, n, repeticiones=3):
    tiempos = []
    for _ in range(repeticiones):
        inicio = time.perf_counter()
        funcion(n)
        tiempos.append(time.perf_counter() - inicio)
    return min(tiempos)
```

## Comparar crecimiento

La función cuadrática crece mucho más rápido, así que usamos tamaños menores.

```python
print("Lineal")
for n in [10_000, 20_000, 40_000, 80_000]:
    print(f"n={n:6d} tiempo={medir(lineal, n):.6f}s")

print("\nCuadrática")
for n in [100, 200, 400, 800]:
    print(f"n={n:6d} tiempo={medir(cuadratica, n):.6f}s")
```

## Normalizar por el coste esperado

Si una función es lineal, `tiempo / n` debería ser relativamente estable. Si es cuadrática, miramos `tiempo / n^2`.

```python
print("Lineal: tiempo / n")
for n in [10_000, 20_000, 40_000, 80_000]:
    t = medir(lineal, n)
    print(f"n={n:6d} t/n={t/n:.12f}")

print("\nCuadrática: tiempo / n^2")
for n in [100, 200, 400, 800]:
    t = medir(cuadratica, n)
    print(f"n={n:6d} t/n^2={t/(n*n):.12f}")
```
