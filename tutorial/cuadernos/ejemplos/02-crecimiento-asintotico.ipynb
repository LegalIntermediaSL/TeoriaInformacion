# Comparación de crecimiento asintótico

Este cuaderno acompaña el módulo de fundamentos matemáticos y el módulo de complejidad computacional.

El objetivo es observar cómo funciones que parecen cercanas para entradas pequeñas se separan drásticamente al crecer `n`.

## Tabla de valores

Comparamos algunos órdenes frecuentes: logarítmico, lineal, cuasilineal, cuadrático y exponencial.

```python
import math


def valores(n):
    return {
        "log2 n": math.log2(n),
        "n": n,
        "n log2 n": n * math.log2(n),
        "n^2": n**2,
        "2^n": 2**n,
    }


tamanos = [2, 4, 8, 16, 32]
columnas = ["log2 n", "n", "n log2 n", "n^2", "2^n"]

print(f"{'n':>5} " + " ".join(f"{c:>14}" for c in columnas))
for n in tamanos:
    fila = valores(n)
    print(f"{n:5d} " + " ".join(f"{fila[c]:14.2f}" for c in columnas))
```

## Mirar el crecimiento con escala logarítmica

La función `2^n` crece tan rápido que una escala lineal suele ocultar a las demás. Para comparar mejor, graficamos el eje vertical en escala logarítmica.

```python
ns = list(range(2, 31))
series = {
    "log2 n": [math.log2(n) for n in ns],
    "n": ns,
    "n log2 n": [n * math.log2(n) for n in ns],
    "n^2": [n**2 for n in ns],
    "2^n": [2**n for n in ns],
}

try:
    import matplotlib.pyplot as plt
except ImportError:
    print("matplotlib no está instalado; se omite la gráfica.")
else:
    plt.figure(figsize=(8, 5))
    for nombre, ys in series.items():
        plt.plot(ns, ys, label=nombre)
    plt.yscale("log")
    plt.xlabel("n")
    plt.ylabel("valor, escala logarítmica")
    plt.title("Comparación de órdenes de crecimiento")
    plt.legend()
    plt.grid(True, which="both", alpha=0.3)
    plt.show()
```

## Presupuesto de operaciones

Supongamos que una máquina puede hacer `100_000_000` operaciones por segundo. ¿Cuánto tardarían distintos costes para varios tamaños de entrada?

```python
operaciones_por_segundo = 100_000_000


def segundos_a_texto(segundos):
    if segundos < 1e-6:
        return "< 1 microsegundo"
    if segundos < 1:
        return f"{segundos * 1000:.3f} ms"
    if segundos < 60:
        return f"{segundos:.3f} s"
    if segundos < 3600:
        return f"{segundos / 60:.2f} min"
    if segundos < 86400:
        return f"{segundos / 3600:.2f} h"
    return f"{segundos / 86400:.2f} días"


for n in [20, 30, 40, 50]:
    print(f"n = {n}")
    for nombre in ["n^2", "2^n"]:
        coste = valores(n)[nombre]
        tiempo = coste / operaciones_por_segundo
        print(f"  {nombre:4s}: {segundos_a_texto(tiempo)}")
```
