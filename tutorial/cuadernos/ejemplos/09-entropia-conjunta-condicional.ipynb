# Entropía conjunta y condicional

Este cuaderno calcula `H(X)`, `H(Y)`, `H(X,Y)` y `H(X|Y)` para distribuciones conjuntas pequeñas.

Veremos tres situaciones: independencia, copia perfecta y relación con ruido.

```python
import math


def entropia(distribucion):
    valor = -sum(p * math.log2(p) for p in distribucion.values() if p > 0)
    return 0.0 if abs(valor) < 1e-12 else valor


def marginal(conjunta, indice):
    resultado = {}
    for par, p in conjunta.items():
        clave = par[indice]
        resultado[clave] = resultado.get(clave, 0) + p
    return resultado


def entropia_conjunta(conjunta):
    return entropia(conjunta)


def entropia_condicional_x_dado_y(conjunta):
    py = marginal(conjunta, 1)
    total = 0
    for y, prob_y in py.items():
        distribucion = {x: p / prob_y for (x, y2), p in conjunta.items() if y2 == y}
        total += prob_y * entropia(distribucion)
    return total
```

## Casos de comparación

Todas las distribuciones usan dos variables binarias.

```python
casos = {
    "independientes": {(0, 0): 0.25, (0, 1): 0.25, (1, 0): 0.25, (1, 1): 0.25},
    "copia perfecta": {(0, 0): 0.5, (0, 1): 0.0, (1, 0): 0.0, (1, 1): 0.5},
    "copia con ruido": {(0, 0): 0.4, (0, 1): 0.1, (1, 0): 0.1, (1, 1): 0.4},
}

for nombre, conjunta in casos.items():
    hx = entropia(marginal(conjunta, 0))
    hy = entropia(marginal(conjunta, 1))
    hxy = entropia_conjunta(conjunta)
    hx_y = entropia_condicional_x_dado_y(conjunta)
    print(f"{nombre:16s} H(X)={hx:.3f} H(Y)={hy:.3f} H(X,Y)={hxy:.3f} H(X|Y)={hx_y:.3f}")
```

## Verificar la regla de la cadena

Debe cumplirse `H(X,Y) = H(Y) + H(X|Y)`.

```python
for nombre, conjunta in casos.items():
    hy = entropia(marginal(conjunta, 1))
    hx_y = entropia_condicional_x_dado_y(conjunta)
    hxy = entropia_conjunta(conjunta)
    print(nombre, round(hxy, 6), round(hy + hx_y, 6))
```
