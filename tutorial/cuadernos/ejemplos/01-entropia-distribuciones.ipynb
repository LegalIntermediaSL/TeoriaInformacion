# Entropía de distribuciones discretas

Este cuaderno acompaña el artículo sobre entropía. Calcula la información propia y la entropía de varias fuentes discretas pequeñas.

La idea que queremos observar es simple: una distribución más uniforme suele tener mayor entropía que una distribución muy concentrada.

## Funciones básicas

Usaremos logaritmos en base 2, así que las unidades serán bits.

```python
import math


def informacion_propia(probabilidad):
    """Devuelve -log2(p), la sorpresa asociada a un resultado."""
    if probabilidad <= 0 or probabilidad > 1:
        raise ValueError("La probabilidad debe estar en el intervalo (0, 1].")
    return -math.log2(probabilidad)


def entropia(distribucion):
    """Calcula la entropía de una distribución discreta dada como diccionario."""
    total = sum(distribucion.values())
    if not math.isclose(total, 1.0, abs_tol=1e-9):
        raise ValueError(f"Las probabilidades deben sumar 1; suman {total}.")
    return sum(p * informacion_propia(p) for p in distribucion.values() if p > 0)
```

## Comparación de fuentes

Comparamos monedas equilibradas, monedas sesgadas y distribuciones uniformes con más símbolos.

```python
distribuciones = {
    "moneda equilibrada": {"cara": 0.5, "cruz": 0.5},
    "moneda sesgada 0.9": {"cara": 0.9, "cruz": 0.1},
    "moneda sesgada 0.99": {"cara": 0.99, "cruz": 0.01},
    "4 símbolos uniformes": {"a": 0.25, "b": 0.25, "c": 0.25, "d": 0.25},
    "fuente concentrada": {"a": 0.7, "b": 0.1, "c": 0.1, "d": 0.1},
}

for nombre, distribucion in distribuciones.items():
    print(f"{nombre:22s} H = {entropia(distribucion):.4f} bits")
```

## Entropía de una moneda según su sesgo

La entropía de una moneda es máxima cuando `P(cara) = 0.5` y disminuye cuando un resultado se vuelve casi seguro.

```python
probabilidades = [i / 100 for i in range(1, 100)]
entropias = [entropia({"cara": p, "cruz": 1 - p}) for p in probabilidades]

p_max = probabilidades[entropias.index(max(entropias))]
print(f"Entropía máxima aproximada: {max(entropias):.4f} bits")
print(f"Se alcanza cerca de P(cara) = {p_max:.2f}")
```

```python
try:
    import matplotlib.pyplot as plt
except ImportError:
    print("matplotlib no está instalado; se omite la gráfica.")
else:
    plt.figure(figsize=(7, 4))
    plt.plot(probabilidades, entropias)
    plt.xlabel("P(cara)")
    plt.ylabel("Entropía H(X) en bits")
    plt.title("Entropía de una moneda sesgada")
    plt.grid(True, alpha=0.3)
    plt.show()
```
