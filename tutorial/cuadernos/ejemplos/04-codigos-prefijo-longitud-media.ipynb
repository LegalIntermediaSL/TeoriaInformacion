# Códigos prefijo y longitud media

Este cuaderno explora una idea central de la codificación de fuente: si los símbolos no son igualmente probables, conviene asignar palabras de código más cortas a los símbolos frecuentes.

Trabajaremos con códigos prefijo, es decir, códigos donde ninguna palabra de código es prefijo de otra.

## Herramientas

Definimos funciones para comprobar si un código es prefijo, calcular su longitud media y evaluar la suma de Kraft.

```python
def es_codigo_prefijo(codigo):
    palabras = list(codigo.values())
    for i, palabra in enumerate(palabras):
        for j, otra in enumerate(palabras):
            if i != j and otra.startswith(palabra):
                return False
    return True


def longitud_media(distribucion, codigo):
    return sum(distribucion[simbolo] * len(codigo[simbolo]) for simbolo in distribucion)


def suma_kraft(codigo):
    return sum(2 ** (-len(palabra)) for palabra in codigo.values())
```

## Una fuente no uniforme

La fuente produce cuatro símbolos, pero `A` aparece con más frecuencia que los demás.

```python
fuente = {"A": 0.5, "B": 0.25, "C": 0.125, "D": 0.125}

codigo_fijo = {"A": "00", "B": "01", "C": "10", "D": "11"}
codigo_variable = {"A": "0", "B": "10", "C": "110", "D": "111"}

for nombre, codigo in [("fijo", codigo_fijo), ("variable", codigo_variable)]:
    print(nombre)
    print("  prefijo:", es_codigo_prefijo(codigo))
    print("  suma de Kraft:", suma_kraft(codigo))
    print("  longitud media:", longitud_media(fuente, codigo))
```

El código variable es mejor para esta fuente porque asigna la palabra más corta al símbolo más probable.

## Codificar y decodificar

Un código prefijo puede decodificarse leyendo de izquierda a derecha sin separadores.

```python
def codificar(mensaje, codigo):
    return "".join(codigo[simbolo] for simbolo in mensaje)


def decodificar(bits, codigo):
    inverso = {palabra: simbolo for simbolo, palabra in codigo.items()}
    resultado = []
    buffer = ""
    for bit in bits:
        buffer += bit
        if buffer in inverso:
            resultado.append(inverso[buffer])
            buffer = ""
    if buffer:
        raise ValueError("La cadena no termina en una palabra de código completa.")
    return resultado


mensaje = list("ABACABAD")
bits = codificar(mensaje, codigo_variable)

print("mensaje original:", mensaje)
print("bits codificados:", bits)
print("decodificado:", decodificar(bits, codigo_variable))
```

## Un código problemático

Si una palabra es prefijo de otra, la decodificación puede ser ambigua.

```python
codigo_ambiguo = {"A": "0", "B": "01", "C": "10", "D": "11"}

print("¿Es prefijo?", es_codigo_prefijo(codigo_ambiguo))
print("Suma de Kraft:", suma_kraft(codigo_ambiguo))
```
