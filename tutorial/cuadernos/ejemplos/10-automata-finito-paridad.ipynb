# Autómata finito para paridad

Este cuaderno implementa un autómata finito determinista que reconoce cadenas binarias con número par de unos.

Es un ejemplo mínimo de cómputo con memoria finita.

```python
automata = {
    "estado_inicial": "par",
    "estados_aceptacion": {"par"},
    "transiciones": {
        ("par", "0"): "par",
        ("par", "1"): "impar",
        ("impar", "0"): "impar",
        ("impar", "1"): "par",
    },
}


def ejecutar(automata, cadena):
    estado = automata["estado_inicial"]
    historia = [estado]
    for simbolo in cadena:
        estado = automata["transiciones"][(estado, simbolo)]
        historia.append(estado)
    acepta = estado in automata["estados_aceptacion"]
    return acepta, historia
```

## Probar cadenas

El autómata acepta si la cantidad de unos es par.

```python
cadenas = ["", "0", "1", "11", "101", "1011", "1111", "10010"]

for cadena in cadenas:
    acepta, historia = ejecutar(automata, cadena)
    print(f"{cadena!r:8s} acepta={acepta} historia={historia}")
```

## Crear otro autómata

Ahora reconocemos cadenas que terminan en `1`.

```python
termina_en_uno = {
    "estado_inicial": "no",
    "estados_aceptacion": {"si"},
    "transiciones": {
        ("no", "0"): "no",
        ("no", "1"): "si",
        ("si", "0"): "no",
        ("si", "1"): "si",
    },
}

for cadena in ["0", "1", "1010", "1011"]:
    acepta, _ = ejecutar(termina_en_uno, cadena)
    print(f"{cadena!r:6s} termina en 1: {acepta}")
```
