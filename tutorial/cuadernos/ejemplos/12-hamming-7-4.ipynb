# Código de Hamming (7,4)

Este cuaderno implementa el código de Hamming (7,4), que codifica 4 bits de datos en 7 bits y corrige un error de un solo bit.

Es una mejora conceptual respecto del código de repetición: usa redundancia de forma más eficiente.

```python
def xor(*bits):
    resultado = 0
    for bit in bits:
        resultado ^= bit
    return resultado


def codificar_hamming_74(datos):
    if len(datos) != 4 or any(bit not in (0, 1) for bit in datos):
        raise ValueError("Se esperan 4 bits de datos.")
    d1, d2, d3, d4 = datos
    p1 = xor(d1, d2, d4)
    p2 = xor(d1, d3, d4)
    p4 = xor(d2, d3, d4)
    return [p1, p2, d1, p4, d2, d3, d4]


def sindrome(palabra):
    b1, b2, b3, b4, b5, b6, b7 = palabra
    s1 = xor(b1, b3, b5, b7)
    s2 = xor(b2, b3, b6, b7)
    s4 = xor(b4, b5, b6, b7)
    return s1 + 2 * s2 + 4 * s4


def corregir(palabra):
    corregida = palabra[:]
    posicion = sindrome(corregida)
    if posicion:
        corregida[posicion - 1] ^= 1
    return corregida, posicion


def decodificar_hamming_74(palabra):
    corregida, posicion = corregir(palabra)
    datos = [corregida[2], corregida[4], corregida[5], corregida[6]]
    return datos, corregida, posicion
```

## Codificar y corregir

Introducimos un error artificial y dejamos que el síndrome localice la posición.

```python
datos = [1, 0, 1, 1]
codigo = codificar_hamming_74(datos)
recibido = codigo[:]
recibido[4] ^= 1

decodificado, corregida, posicion = decodificar_hamming_74(recibido)

print("datos:      ", datos)
print("codificado: ", codigo)
print("recibido:   ", recibido)
print("síndrome:   ", sindrome(recibido))
print("posición:   ", posicion)
print("corregida:  ", corregida)
print("decodificado:", decodificado)
```

## Probar todos los errores de un bit

Un código Hamming (7,4) corrige cualquier error individual.

```python
codigo = codificar_hamming_74([1, 1, 0, 1])

for i in range(7):
    recibido = codigo[:]
    recibido[i] ^= 1
    datos, corregida, posicion = decodificar_hamming_74(recibido)
    print(f"error en posición {i + 1}: síndrome={posicion}, correcto={corregida == codigo}, datos={datos}")
```
